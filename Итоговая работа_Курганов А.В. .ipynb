### Столбцы данных
- id - ID наблюдения в рамках набора данных. Единица наблюдения в датасете — совокупность личных дел с соответствующими характеристиками.<br>
- district — Федеральный округ.<br>
- region - Регион.<br>
- month_application - Месяц обращения в орган занятости населения. Для Чеченской Республики в исходных данных с мая 2019 года отсутствовали сведения о дате признания гражданина безработным. Эта дата была восстановлена следующим образом: для всех наблюдений, у которых была указана дата последней перерегистрации в качестве безработного, датой признания гражданина безработным может считаться дата обращения гражданина в орган занятости населения. Такой подход соответствует порядку, установленному правилами регистрации безработных граждан, и позволяет с достаточно высокой точностью восстановить пропущенные наблюдения. Для восстановленных таким образом дат атрибут changed_value принимает значение 1.<br>
- changed_value - отметка о том, что дата регистрации в качестве безработного была восстановлена по косвенным признакам.<br>
- age — Группа по возрасту, к которой относится безработный гражданин. Принимает значения: 16-21; 22-27; 28-33; 34-39; 40-45; 46-51; 52-57; 58-60.<br>
- gender — Пол безработного: 1 — мужской; 0 — женский. <br>
- education — Уровень образования безработного: 1 — начальное общее образование; 2 — основное общее образование; 3 — среднее общее образование; 4 — среднее профессиональное образование; 6 — высшее образование 9 — другое. Коды приведены в соответствии с Международной стандартной классификацией образования. В категорию «высшее образование» (код 6) включается высшее образование, полученное в рамках специалитета, бакалавриата, магистратуры, подготовки кадров высшей квалификации.<br>
- hc_repeat — отметка о том, что безработный обратился повторно в органы занятости населения.<br>
- hc_didnotwork — отметка о том, что, по информации органов занятости населения, гражданин ранее официально не работал.<br>
- find_new_job отметка о том, что гражданин нашел работу, в какой-либо период после признания безработным (в случае, если причина закрытия личного дела — «Трудоустройство» или была указана дата трудоустройства). Органы занятости отслеживают статус занятости безработных только до тех пор, пока на них открыто личное дело. При этом личное дело может быть закрыто не только в связи с трудоустройством, но и по другим причинам: например, прохождение профессионального обучения или длительная неявка в ЦЗН (причины закрытия личных дел содержатся в датасете с микроданными в продвинутом режиме доступа]). Кроме того, некоторым безработным, которые, например, обратились в ЦЗН во второй половине 2020 года, требуется больше времени на поиск новой работы (выгрузка данных осуществлялась в конце декабря 2020 года). Таким образом, в данных оценка доли трудоустроенных безработных занижена — данные цензурированы справа, необходимо использовать специальные методы для учета этой особенности.<br>
- szn_name - наименование органа занятости, в который обратился гражданин.<br>
- oktmo - ОКТМО муниципального района/городского округа, в котором расположен орган занятости, в который обратился гражданин.<br>
- latitude_dd - широта органа занятости, в который обратился гражданин, в формате DD (градусы с десятичной частью).<br>
- longitude_dd - долгота органа занятости, в который обратился гражданин, в формате DD (градусы с десятичной частью).<br>
- count - количество личных дел с соответствующими характеристиками.<br>

```python
URL = 'https://ru.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D1%81%D1%83%D0%B1%D1%8A%D0%B5%D0%BA%D1%82%D0%BE%D0%B2_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B9%D1%81%D0%BA%D0%BE%D0%B9_%D0%A4%D0%B5%D0%B4%D0%B5%D1%80%D0%B0%D1%86%D0%B8%D0%B8'
data2 = pd.read_html(URL, header=0, decimal  = ',')[0] # загружаем с указанного выше адреса первую таблицу на странице. 
data2.head()
```

#  <a name="4">4. Проверка гипотез </a>

* H0: Снятие с учёта в органах занятости населения, в связи с нахождением новой работы не зависит от образования.
* H1: Снятие с учёта в органах занятости населения, в связи с нахождением новой работы зависит от образования.
<br>
<br>

* H0: Повторное обращение в органы занятости населения не зависит от уровня образования граждан.
* H1: Повторное обращение в органы занятости населения не зависит от уровня образования граждан.
<br>
<br>

**H0: Снятие с учёта в органах занятости населения, в связи с нахождением новой работы не зависит от уровня образования.**<br>
**H1: Снятие с учёта в органах занятости населения, в связи с нахождением новой работы зависит от уровня образования.**

Для начала, разобъём записи в датафрейме на 2 группы: граждане с высшим образованием и граждане с общим образованием.

```python
# Фильтруем датафрейм по высшему и среднему образованию
new_df = result_df.loc[((result_df['education'] == 'Высшее образование') | (result_df['education'] == 'Общее образование'))]

# Заменяем значения высшего образования на True, а среднего - на False
new_df['education'] = new_df['education'].replace({'Высшее образование': True, 'Общее образование': False})
```

На основе созданных выборок, сделаем таблицу сопряжённости:

```python
hyp_table = pd.crosstab(new_df['education'], new_df['find_new_job'], values = new_df['count'], aggfunc = 'sum')
```

```python
# Производим Хи-квадрат тест
chi2, p, dof, expected = chi2_contingency(hyp_table)

alpha = 0.05

print("p-value:", p)
print("Хи-квадрат статистика:", chi2)

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Принимаем нулевую гипотезу")
```

```python
hyp_table.plot(figsize = (8,6), kind = 'bar')
plt.xlabel("Образование (True - высшее; False - общее)", fontsize = 15)
plt.ylabel("Количество обращений в ЦЗН", fontsize = 15) 
plt.figtext(0.5, -0.3, "Рисунок 5. - График зависимости снятия с учёта в связи с нахождением работы и уровнем образования.",fontsize = 15)
plt.show()
```

**В ходе проверки гипотезы, было установлено, что есть зависимость между уровнем образования гражданина, вставшего на учёт в качестве безработного и снятие его с данного учёта, в связи с нахождением новой работы. p-value = 0. Это означает, что следует отвергнуть нулевую гипотезу о том, что исследуемые данные независимы друг от друга. Высокая статитстика Хи-квадрат = 85994.60 показывает, что есть статистически значимая разница между гражданами с высшим и средним образованием, снявшимися с учёта ЦЗН в связи с нахождением новой работы.<br>
Как видно из графика выше, обе группы граждан чаще снимаются с учёта в органах занятости не  связи с нахождением новой работы. Однако, у граждан с высшим образованием соотношение снявшихся с учёта в связи с нахождением новой работы гораздо выше, чем у граждан, имеющих только общее образование.** 

<br>
<br>

**H0: Повторное обращение в органы занятости населения не зависит от уровня образования граждан.<br>**
**H1: Повторное обращение в органы занятости населения зависит от уровня образования граждан.**

На основе созданных выборок, сделаем таблицу сопряжённости:

```python
hyp_table2 = pd.crosstab(new_df['education'], new_df['hc_repeat'], values = new_df['count'], aggfunc = 'sum')
```

```python
# Производим Хи-квадрат тест
chi2, p, dof, expected = chi2_contingency(hyp_table2)

alpha = 0.05

print("p-value:", p)
print("Хи-квадрат статистика:", chi2)

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Принимаем нулевую гипотезу")
```

```python
hyp_table2.plot(figsize = (8,6), kind = 'bar')
plt.xlabel("Образование (True - высшее; False - общее)", fontsize = 15)
plt.ylabel("Количество обращений в ЦЗН", fontsize = 15) 
plt.figtext(0.5, -0.3, "Рисунок 6. - График зависимости повторного обращения в орагны занятости населения и уровнем образования.",fontsize = 15)
plt.show()
```

**В ходе проверки гипотезы, было установлено, что есть зависимость между уровнем образования гражданина, вставшего на учёт в качестве безработного и снятие его с данного учёта, в связи с нахождением новой работы. p-value = 0. Это означает, что следует отвергнуть нулевую гипотезу о том, что исследуемые данные независимы друг от друга. Высокая статитстика Хи-квадрат = 16226.21 показывает, что есть статистически значимая разница между гражданами с высшим и средним образованием, повторно обращающимися в ЦЗН.<br>

Чаще других, повтороно регистрируются в качестве безработных женщины этой же возрастной группы, однако, имеющие только среднее образование. В ходе проведённого анализа, было установлено, что количество повторных обращений в ЦЗН и снятие с учёта в органах трудоустройства населения взаимосвязаны с уровнем образования граждан, а из приведённых выше графиков следует, что граждане с высшим образованием реже повторно регистрируются в качестве безработных и чаще снимаются с учёта в органах занятости населения, в связи с нахождением новой работы.
